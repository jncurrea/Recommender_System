```python
# Import the libraries needed to convert to dataframe and vectorize it
import pandas as pd
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
# import spacy

# Intalize input file to be used for the couunt list
input_file = "0.translated_test.csv"
```

```python
# Create Data Frame and return column names to know what they are called
df = pd.read_csv(input_file)
df.columns

# Know which column name you will need to change to run analysis
```

```python

# Download necessary NLTK data files
nltk.download('stopwords')

# Get stopwords from NLTK
stop_words = set(stopwords.words('english'))

def preprocess(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove number sequences (e.g., "123", "4567") and non-English characters (keep ASCII)
    text = re.sub(r'\d+', '', text)        # Remove digits
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)  # Remove non-ASCII characters
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation

    # Tokenize and Remove stopwords 
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply preprocessing to the text column
df['clean_text'] = df['message'].apply(preprocess)
```

```python

# Vectorize the data
# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the text data
X = vectorizer.fit_transform(df['clean_text'])

# Convert the result to a DataFrame for better readability (optional)
vectorized_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

vectorized_df

```
